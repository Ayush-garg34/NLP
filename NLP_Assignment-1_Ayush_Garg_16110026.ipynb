### Downloading the file

```python
from nltk.tokenize import word_tokenize
import urllib

url = "http://www.gutenberg.org/files/74/74-0.txt"

# tom = open('tom.txt','r')
tomSawyer = urllib.urlopen(url)
corpus = tomSawyer.read().decode('utf-8')
# corpus = tom.read()
# print corpus
```

```python
import nltk
nltk.download('punkt')
```

### Finding Tokens,Types and TTR

```python
tokens = word_tokenize(corpus)
types={}

# token_number=0
for word in tokens:
    word=word.lower()
    if word in types:
        count = types.get(word,0)
        types[word]=count+1
    else:
        types[word]=1
```

```python
ttr = len(types)/(1.0*len(tokens))
print ("number of Types = "+str(len(types)))
print ("number of Token = "+str(len(tokens)))
print ("TTR = "+ str(ttr))


```

### Zipf's Law

```python
types_sort=sorted(types,key = types.get,reverse = True)

rank=[]
i=0
for word in types_sort:
    rank.insert(i,types.get(word,0))
    i=i+1

# print types_sort  

# frequency={}
# for word in types_sort:
#     frequency[word]=types.get(word,0)
# print frequency

```

```python
print ("|Rank(r)\t|Frequency(f)\t|f.r|")
print ("\n")
index=1

rank_points=[]
frequency_points=[]

for i in rank:
    if i in frequency_points:
        index=index-1
    else:
        rank_points.append(index)
        frequency_points.append(i)
    print( str(index)+"\t"+"\t" +str(i)+ "\t"+"\t"+str(index*i))
    index=index+1
    

# print rank_points
# print frequency_points
```

### Frequency Vs Length

```python
length_frequency={}
length_number_of_instance={}

for i in types:
    if len(i) in length_frequency:
        length_frequency[len(i)]=length_frequency[len(i)]+types.get(i,1)
        length_number_of_instance[len(i)]=length_number_of_instance[len(i)]+1;
    else:
        length_frequency[len(i)]=types.get(i,1)
        length_number_of_instance[len(i)]=1;

        
for i in length_frequency:
    length_frequency[i]=length_frequency[i]/length_number_of_instance[i]
# print length_frequency
# print length_number_of_instance
        
```

```python
Length=[]
Freq=[]
for i in  length_frequency:
    Length.append(i)
    Freq.append(length_frequency.get(i,0))
    
plt.plot(Freq,Length)
plt.xlabel('Frequency')
plt.ylabel('Length')
plt.title('Frequency Vs Length Plot')
plt.show()
  



```

### Heaps Law

```python
from nltk.tokenize import word_tokenize
import urllib
url = "http://www.gutenberg.org/files/74/74-0.txt"
tom=urllib.urlopen(url)
corpus2=tom.readlines()

V=[]
N=[]
v=""
counter=0;
for i in corpus2:
    counter=counter+1
    v=v+i.decode('utf-8')
    if(counter%25==0):
        Tokens=word_tokenize(v)
        Types={}
        for word in Tokens:
            word=word.lower()
            if word in Types:
                count = Types.get(word,0)
                Types[word]=count+1
            else:
                Types[word]=1
        V.append(len(Types))
        N.append(len(Tokens))

        
```

```python
plt.plot(N,V)
plt.ylabel('Size of Vocabulary')
plt.xlabel('Number of Tokens')
plt.title('Heaps Law')
plt.show()
```

```python
# Reference
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html


def func(n, k,b):
    return k*(n**b)

from scipy.optimize import curve_fit
a,b=curve_fit(func, N, V)
K=a[0]
Beta =a[1] 
print ("K = " +str(K))
print ("Beta = "+str(Beta))
```

```python
plt.plot(N,func(N,K,Beta),'y')
plt.plot(N,V)
plt.ylabel('Size of Vocabulary')
plt.xlabel('Number of Tokens')
plt.title('Heaps Law')

plt.show()
```
